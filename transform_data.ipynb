```python
import json

with open("/content/sample_data/2_대화체.json", 'r') as f :
  data = json.load(f)

data
```

# Llama2 학습을 위한 프롬프트 형태

```python
data_dict = {}
data_num = 200

for i in range(data_num) :
  kor = data[i]["한국어"]
  eng = data[i]["영어"]
  data_dict[kor] = eng

data_dict
```

```python
final_list = []

for idx, (kor, eng) in enumerate(data_dict.items()) :
  prompt = f"Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.\
### Instruction : Korean and English sentences are given in order below. Translate Korean sentences into English sentences.\
### Korean : {kor} ### English : {eng}"
  print(idx, prompt)
  prompt_dict = {}
  prompt_dict["text"] = prompt
  final_list.append(prompt_dict)
```

```python
final_list
```

```python
with open("conversational.json", 'w', encoding = "utf-8") as f :
  f.write("[\n")
  for i, item in enumerate(final_list) :
    json_str = json.dumps(item, ensure_ascii = False)
    f.write(json_str)
    if i < len(final_list) - 1 :
      f.write(",\n")
    else :
      f.write("\n")
  f.write("]\n")
```
